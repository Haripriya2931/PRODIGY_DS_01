```python
# Histogram for math score in the dataset
# Continuous variable

plt.figure(figsize=(8,5))
plt.hist(data["math_score"], bins=4,color='pink',edgecolor='purple')
plt.xlabel("score")
plt.ylabel("Average number of students")
plt.show()
```

```python
# Bar chart for Gender in the dataset
# Categorical variable

plt.figure(figsize = (8,5))
data['gender'].value_counts().plot(kind='bar', color =['yellow','lightblue'],edgecolor='black')
plt.xlabel("Gender")
plt.ylabel("Frequency")
plt.show()
```

```python
# Bar chart for extracurricular_activities in the dataset
# Categorical variable

plt.figure(figsize = (8,5))
data['extracurricular_activities'].value_counts().plot(kind='bar', color =['salmon','lightgreen','yellow'],edgecolor='black')
plt.xlabel("extracurricular_activities")
plt.ylabel("Frequency")
plt.show()
```

```python
# Bar chart for part_time_job in the dataset
# Categorical variable

plt.figure(figsize = (8,5))
data['part_time_job'].value_counts().plot(kind='bar', color =['blue','royalblue'],edgecolor='black')
plt.xlabel("part_time_job")
plt.ylabel("Frequency")
plt.show()
```

```python
# Histogram for study time weekly of the students in the dataset
# Continuous variable

plt.figure(figsize=(8,5))
plt.hist(data["weekly_self_study_hours"], bins=100,color='mediumslateblue',edgecolor='black')
plt.xlabel("Study time weekly of the Students")
plt.ylabel("Frequency")
plt.show()
```

```python
# Histogram for absence_days of the Students in the dataset
# Continuous variable

plt.figure(figsize=(10,5))
plt.hist(data["absence_days"], bins=30,color='orange',edgecolor='black')
plt.xlabel("absence_days of the Students")
plt.ylabel("Frequency")
plt.show()
```
